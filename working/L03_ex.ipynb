### Q2 : Planetary positions

```python
# semi-major axes
a = np.array([0.39, 0.72, 1.00, 1.52, 5.20, 9.54, 19.22, 30.06, 39.48])

# orbital periods
P = np.array([0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8, 248.09])

# planet names
names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", 
         "Uranus", "Neptune", "Pluto"]
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlim(2*10**(-1),10**2)

ax.scatter(a, P)
ax.set_xlabel("semi-major axis (AU)")
ax.set_ylabel("orbital period (yr)")

for i in range(len(a)):
    ax.text(a[i]+a[i]/7,P[i],names[i],va='center')
```

### Q3 : A perfect circle

```python
theta = np.linspace(0, 2*np.pi, 100)
R = 1
x = R*np.cos(theta)
y = R*np.sin(theta)
```

```python
fig = plt.figure()
ax = fig.add_subplot(aspect=1)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.plot(x,y,'green')
ax.fill(x, y, 'green',alpha=0.5);
```

```python
def draw_circle(x0, y0, R, color):
    theta = np.linspace(0, 2*np.pi, 100)
    x = x0 + R*np.cos(theta)
    y = y0 + R*np.sin(theta)
    ax.fill(x,y, color,alpha=0.5)
```

```python
fig = plt.figure()
ax = fig.add_subplot(aspect=1)

draw_circle(2,3,1,"green")
```

```python
num_circles = 10

x0 = np.random.rand(num_circles)
y0 = np.random.rand(num_circles)
R = 0.3*np.random.rand(num_circles)
```

```python
def draw_circle2(x0, y0, R):
    theta = np.linspace(0, 2*np.pi, 100)
    x = x0 + R*np.cos(theta)
    y = y0 + R*np.sin(theta)
    ax.fill(x,y,alpha=0.5)

dc2 = np.vectorize(draw_circle2)
```

```python
fig = plt.figure()
ax = fig.add_subplot(aspect=1)

_ = dc2(x0,y0,R)
```
